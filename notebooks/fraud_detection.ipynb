</font>

```python
train_data.describe()
```

```python
# do some preprocessing !
class0 = train_data[train_data.Class == 0]
Q1 = class0.quantile(0.25)
Q3 = class0.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
class0 = class0[~((class0 < lower_bound) | (class0 > upper_bound)).any(axis=1)]
```

```python
train_data_new = pd.DataFrame()
train_data_new = pd.merge(class0, train_data[train_data.Class == 1], how='outer')
train_data_new
```

|<code>C</code>|همانطور که گفتیم، هایپرپارامتر همواره مثبت ‍`C` در الگوریتم SVM در Sklearn، عملیات موازنه را انجام می‌دهد. هرچه مقدار این هایپرپارامتر بیشتر باشد، اهمیت Hard Margin بیشتر خواهد شد. به عبارت دیگر، این آرگومان در منظم‌سازی موثر است. البته تاثیر `C` در منظم‌سازی، به صورت معکوس است. به این معنی که با کاهش `C`، الگوریتم عبارت‌های مخصوص منظم‌سازی، اثر بیشتری بر ضرایب خواهد داشت. |

    در اینجا با توجه به imbalance بودن دیتاست و نامتوازن‌بودن تعداد از هر نوع کلاس، معیاری که برای ارزیابی عملکرد مدل انتخاب کرده‌ایم، <code>F1-score</code> نام دارد.

    این معیار، سنجه ارزیابی کیفیت مدل شماست. به عبارت بهتر در سامانه داوری هم از همین معیار برای نمره‌دهی استفاده شده است.

```python
# evaluate your model
from sklearn.metrics import f1_score, classification_report
from sklearn.model_selection import train_test_split

model = SVC(C=1, kernel='linear')

X_train, X_test, y_train, y_test = train_test_split(train_data_new.drop(columns='Class'), train_data_new.Class, test_size=0.2, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the F1 score for binary classification
f1 = f1_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)
```
